```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("addhealth_pds.csv",low_memory=False)
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
data['H1PF29'] = pd.to_numeric(data['H1PF29'])
data['H1PF30']=pd.to_numeric(data['H1PF30'])
data['H1PF31']=pd.to_numeric(data['H1PF31'])
data['H1PF32']=pd.to_numeric(data['H1PF32'])
data['H1PF33']=pd.to_numeric(data['H1PF33'])
data['H1PF34']=pd.to_numeric(data['H1PF34'])
data['H1PF35']=pd.to_numeric(data['H1PF35'])
data['H1PF36']=pd.to_numeric(data['H1PF36'])
data['H1ED20']=pd.to_numeric(data['H1ED20'])
data['H1ED18']=pd.to_numeric(data['H1ED18'])
 
```

```python
data2=data.copy()
data2['H1PF30']=data2['H1PF30'].replace(3,np.nan)
data2['H1PF30']=data2['H1PF30'].replace(6,np.nan)
data2['H1PF30']=data2['H1PF30'].replace(8,np.nan)

data2['H1PF32']=data2['H1PF32'].replace(3,np.nan)
data2['H1PF32']=data2['H1PF32'].replace(6,np.nan)
data2['H1PF32']=data2['H1PF32'].replace(8,np.nan)

data2['H1PF33']=data2['H1PF33'].replace(3,np.nan)
data2['H1PF33']=data2['H1PF33'].replace(6,np.nan)
data2['H1PF33']=data2['H1PF33'].replace(8,np.nan)

data2['H1PF34']=data2['H1PF34'].replace(3,np.nan)
data2['H1PF34']=data2['H1PF34'].replace(6,np.nan)
data2['H1PF34']=data2['H1PF34'].replace(8,np.nan)

data2['H1PF35']=data2['H1PF35'].replace(3,np.nan)
data2['H1PF35']=data2['H1PF35'].replace(6,np.nan)
data2['H1PF35']=data2['H1PF35'].replace(8,np.nan)

data2['H1PF36']=data2['H1PF36'].replace(3,np.nan)
data2['H1PF36']=data2['H1PF36'].replace(6,np.nan)
data2['H1PF36']=data2['H1PF36'].replace(8,np.nan)
```

```python
data2['H1ED18']=data2['H1ED18'].replace(3,np.nan)
data2['H1ED18']=data2['H1ED18'].replace(6,np.nan)
data2['H1ED18']=data2['H1ED18'].replace(7,np.nan)
data2['H1ED18']=data2['H1ED18'].replace(8,np.nan)

data2['H1ED20']=data2['H1ED20'].replace(3,np.nan)
data2['H1ED20']=data2['H1ED20'].replace(6,np.nan)
data2['H1ED20']=data2['H1ED20'].replace(7,np.nan)
data2['H1ED20']=data2['H1ED20'].replace(8,np.nan)
```

```python
recode={1:5,2:4,4:2,5:1}
```

```python
data2['H1PF30FREQ']=data['H1PF30'].map(recode)
data2['H1PF32FREQ']=data['H1PF32'].map(recode)
data2['H1PF33FREQ']=data['H1PF33'].map(recode)
data2['H1PF34FREQ']=data['H1PF34'].map(recode)
data2['H1PF35FREQ']=data['H1PF35'].map(recode)
data2['H1PF36FREQ']=data['H1PF36'].map(recode)
data2['H1ED18FREQ']=data['H1ED18'].map(recode)
data2['H1ED20FREQ']=data['H1ED20'].map(recode)
```

```python
data2['H1PF30FREQ']=data2['H1PF30FREQ'].astype('category')
data2['H1PF32FREQ']=data2['H1PF32FREQ'].astype('category')
data2['H1PF33FREQ']=data2['H1PF33FREQ'].astype('category')
data2['H1PF34FREQ']=data2['H1PF34FREQ'].astype('category')
data2['H1PF35FREQ']=data2['H1PF35FREQ'].astype('category')
data2['H1PF36FREQ']=data2['H1PF36FREQ'].astype('category')

data2['H1ED18FREQ']=data2['H1ED18FREQ'].astype('category')
data2['H1ED20FREQ']=data2['H1ED20FREQ'].astype('category')
```

```python
def H1PF30FREQResponse(row):
    if row['H1PF30FREQ']==4 or row['H1PF30FREQ']==5:
        return 1
    else:
        return 0
def H1PF32FREQResponse(row):
    if row['H1PF32FREQ']==4 or row['H1PF32FREQ']==5:
        return 1
    else:
        return 0
    
def H1PF33FREQResponse(row):
    if row['H1PF33FREQ']==4 or row['H1PF33FREQ']==5:
        return 1
    else:
        return 0
    
def H1PF34FREQResponse(row):
    if row['H1PF34FREQ']==4 or row['H1PF34FREQ']==5:
        return 1
    else:
        return 0
def H1PF36FREQResponse(row):
    if row['H1PF36FREQ']==4 or row['H1PF36FREQ']==5:
        return 1
    else:
        return 0 
```

```python
data2['H1PF30FREQResponse']=data2.apply(lambda row:H1PF30FREQResponse(row),axis=1)
data2['H1PF32FREQResponse']=data2.apply(lambda row:H1PF32FREQResponse(row),axis=1)
data2['H1PF33FREQResponse']=data2.apply(lambda row:H1PF33FREQResponse(row),axis=1)
data2['H1PF34FREQResponse']=data2.apply(lambda row:H1PF34FREQResponse(row),axis=1)
data2['H1PF36FREQResponse']=data2.apply(lambda row:H1PF36FREQResponse(row),axis=1)
```
