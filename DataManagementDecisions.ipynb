# Making Data Management Decisions

Importing Libraries

```python
import pandas as pd
import numpy as np
```

Reading CSV

```python
data=pd.read_csv("addhealth_pds.csv",low_memory=False)
```

Converting Variables into Numeric

```python
data['H1PF29'] = pd.to_numeric(data['H1PF29'])
data['H1PF30']=pd.to_numeric(data['H1PF30'])
data['H1PF31']=pd.to_numeric(data['H1PF31'])
data['H1PF32']=pd.to_numeric(data['H1PF32'])
data['H1PF33']=pd.to_numeric(data['H1PF33'])
data['H1PF34']=pd.to_numeric(data['H1PF34'])
data['H1PF35']=pd.to_numeric(data['H1PF35'])
data['H1PF36']=pd.to_numeric(data['H1PF36'])
data['H1ED20']=pd.to_numeric(data['H1ED20'])
data['H1ED18']=pd.to_numeric(data['H1ED18'])
 
```

Copying data

```python
data2=data.copy()
```

Dropping some columns

```python
data2=data2.drop(['H1PF29','H1PF31'],axis=1)
```

Since for variables H1PF30, H1PF32, H1PF33, H1PF34, H1PF35, H1PF36, 6 meaning refused,8 meaning donâ€™t know, 3 meaning neither agree nor disagree are not useful so they will be filled with NaN

```python
data2['H1PF30']=data2['H1PF30'].replace(3,np.nan)
data2['H1PF30']=data2['H1PF30'].replace(6,np.nan)
data2['H1PF30']=data2['H1PF30'].replace(8,np.nan)

data2['H1PF32']=data2['H1PF32'].replace(3,np.nan)
data2['H1PF32']=data2['H1PF32'].replace(6,np.nan)
data2['H1PF32']=data2['H1PF32'].replace(8,np.nan)

data2['H1PF33']=data2['H1PF33'].replace(3,np.nan)
data2['H1PF33']=data2['H1PF33'].replace(6,np.nan)
data2['H1PF33']=data2['H1PF33'].replace(8,np.nan)

data2['H1PF34']=data2['H1PF34'].replace(3,np.nan)
data2['H1PF34']=data2['H1PF34'].replace(6,np.nan)
data2['H1PF34']=data2['H1PF34'].replace(8,np.nan)

data2['H1PF35']=data2['H1PF35'].replace(3,np.nan)
data2['H1PF35']=data2['H1PF35'].replace(6,np.nan)
data2['H1PF35']=data2['H1PF35'].replace(8,np.nan)

data2['H1PF36']=data2['H1PF36'].replace(3,np.nan)
data2['H1PF36']=data2['H1PF36'].replace(6,np.nan)
data2['H1PF36']=data2['H1PF36'].replace(8,np.nan)
```

Since for variables H1ED18, H1ED20, 3,6,7,8 means neither agree nor disagree, refusion, legitimate skip and dont't know, we will convert them into NAN

```python
data2['H1ED18']=data2['H1ED18'].replace(3,np.nan)
data2['H1ED18']=data2['H1ED18'].replace(6,np.nan)
data2['H1ED18']=data2['H1ED18'].replace(7,np.nan)
data2['H1ED18']=data2['H1ED18'].replace(8,np.nan)

data2['H1ED20']=data2['H1ED20'].replace(3,np.nan)
data2['H1ED20']=data2['H1ED20'].replace(6,np.nan)
data2['H1ED20']=data2['H1ED20'].replace(7,np.nan)
data2['H1ED20']=data2['H1ED20'].replace(8,np.nan)
```
