Since we are studying impact of social acceptance and school adjustment(for second question) on self seteem, our independent explanatory variables are H1PF35FREQ, H1ED18FREQ,H1ED20FREQ, our response variables are questions answered on self esteem, H1PF30FREQ, H1PF32FREQ, H1PF33FREQ, H1PF34FREQ, H1PF36FREQ
Here, we will study association of each explanatory variable with each response variable.

Here, in our respone variables, more than two categories are present, so we need to collapse them into two agree,
disagree, agree meaning 1 and diagree meaning 0.

```python
def H1PF30FREQResponse(row):
    if row['H1PF30FREQ']==4 or row['H1PF30FREQ']==5:
        return 1
    else:
        return 0
def H1PF32FREQResponse(row):
    if row['H1PF32FREQ']==4 or row['H1PF32FREQ']==5:
        return 1
    else:
        return 0
    
def H1PF33FREQResponse(row):
    if row['H1PF33FREQ']==4 or row['H1PF33FREQ']==5:
        return 1
    else:
        return 0
    
def H1PF34FREQResponse(row):
    if row['H1PF34FREQ']==4 or row['H1PF34FREQ']==5:
        return 1
    else:
        return 0
def H1PF36FREQResponse(row):
    if row['H1PF36FREQ']==4 or row['H1PF36FREQ']==5:
        return 1
    else:
        return 0 
```

```python
data2['H1PF30FREQResponse']=data2.apply(lambda row:H1PF30FREQResponse(row),axis=1)
data2['H1PF32FREQResponse']=data2.apply(lambda row:H1PF32FREQResponse(row),axis=1)
data2['H1PF33FREQResponse']=data2.apply(lambda row:H1PF33FREQResponse(row),axis=1)
data2['H1PF34FREQResponse']=data2.apply(lambda row:H1PF34FREQResponse(row),axis=1)
data2['H1PF36FREQResponse']=data2.apply(lambda row:H1PF36FREQResponse(row),axis=1)
```
